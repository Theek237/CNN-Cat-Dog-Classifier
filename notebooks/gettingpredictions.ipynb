```python
from keras.models import load_model

model = load_model('/content/drive/MyDrive/Colab Notebooks/CNN Cat and Dog Classifier/models/catdogclassifier.keras')
```

```python
import os
import cv2
import numpy as np

img_path = '/content/drive/MyDrive/Colab Notebooks/CNN Cat and Dog Classifier/data/test_set/test_set/dogs/dog.4254.jpg'

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (100, 100))
img = img/255

img = img.reshape(1,100,100,1)
```

```python
prob = float(model.predict(img))

if prob >= 0.5:
    print(f"Cat ðŸ± ({prob*100:.2f}%)")
else:
    print(f"Dog ðŸ¶ ({(1-prob)*100:.2f}%)")


```
