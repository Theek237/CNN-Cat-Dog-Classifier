```python
from keras.models import load_model

model = load_model('Downloads/archive (1)/catdogclassifier.keras')
```

```python
import os
import numpy as np
import cv2

test_img_folder_path = 'Downloads/archive (1)/train/mix'
category_dict = {0: 'Dog', 1: 'Cat'}  # 0=dog, 1=cat

images = os.listdir(test_img_folder_path)

for image in images:
    img_path = os.path.join(test_img_folder_path, image)

    try:
        original = cv2.imread(img_path)
        gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
        resized = cv2.resize(gray, (100, 100))

        input_img = resized / 255.0
        input_img = input_img.reshape(1, 100, 100, 1)

        prob = float(model.predict(input_img))

        # sigmoid output = probability of class 1 (cat)
        if prob >= 0.5:
            label = 1
            confidence = prob
        else:
            label = 0
            confidence = 1 - prob

        category = category_dict[label]

        

        display_img = cv2.resize(original, (400, 400))

        cv2.putText(display_img, category, (20, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 2)

        cv2.putText(display_img, f"{confidence*100:.2f}%", (20, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 2)

        cv2.imshow('Prediction', display_img)

        if cv2.waitKey(0) == 27:
            break

    except Exception as e:
        print(e, img_path)

cv2.destroyAllWindows()
```
